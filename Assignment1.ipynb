```python
# 1
item1_price = int(input("Enter the price of item 1: "))
q1=int (input('quantity of item one'))
item2_price = int(input("Enter the price of item 2: "))
q2=int (input('quantity of item two'))
total_bill=item1_price*q1+item2_price*q2
print('total bill:',total_bill)
```

```python
# 2
C=int(input("Enter the temperature in celsius"))
F=(C*9/5)+32
print("temperature in fahrenheit:",F)
```

```python
# 3
basic=int(input("Enter the basic salary"))
HRA=0.2*basic
DA=0.1*basic
total_salary=basic+HRA+DA
print("print the net salary:",total_salary)
```

```python
# 4
principle=int(input("Enter the principle:"))
rate_of_interest=int(input("Enter the rate of interest:"))
time=int(input("Enter the time(in years):"))
simple_interest=(principle*rate_of_interest*time)/100
print("simple interest:",simple_interest)
```

```python
# 5
total_min=int(input("Enter the total minutes:"))
hours=total_min//60
minutes=total_min%60
print(hours,"hours and",minutes,"minutes")
```

```python
# 6
weight=int(input("Enter the weight in kg:"))
height=float(input("Enter the height in meters:"))
BMI=weight/(height*height)
print("BMI:",BMI)
```

```python
# 8
a=float(input("Enter side a:"))
b=float(input("Enter side b:"))
c=float(input("Enter side c:"))
s=(a+b+c)/2
area=(s*(s-a)*(s-b)*(s-c))**0.5
print("Area of triangle:",area)
```

```python
# 9
painter_rate = 25 # ₹ per square meter

wall_length = float(input("Enter the wall length in meters: "))
wall_height = float(input("Enter the wall height in meters: "))
wall_area = wall_length * wall_height
window_length = 1.5
window_height = 1.2
window_area = window_length * window_height
total_windows_area = 2 * window_area
area_to_be_painted = wall_area - total_windows_area
total_cost = area_to_be_painted * painter_rate
print(f"Total area to be painted: {area_to_be_painted:.2f} square meters")
print(f"Total cost to paint: ₹{total_cost:.2f}")
```
