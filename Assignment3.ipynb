#1 Write a Python function to Ô¨Ånd the Max of three numbers

```python
def find_max_three_numbers(a, b, c):
    """Finds the maximum of three numbers."""
    return max(a, b, c)

# Example usage:
print(f"Max of (10, 20, 5): {find_max_three_numbers(10, 20, 5)}")
print(f"Max of (0, -1, -10): {find_max_three_numbers(0, -1, -10)}")
```

#2 Write a Python function to sum all the numbers in a list.

```python
def sum_list_numbers(numbers):
    """Sums all numbers in a list."""
    return sum(numbers)

# Example usage:
my_list = [1, 2, 3, 4, 6]
print(f"Sum of {my_list}: {sum_list_numbers(my_list)}")

another_list = [10, -2, 9]
print(f"Sum of {another_list}: {sum_list_numbers(another_list)}")
```

#3  Python program to insert a number to any position in a list (Get the number to be inserted and position from the user)  

```python
my_list = [10, 20, 30, 40, 50]
print(f"Original list: {my_list}")

number = int(input("Enter number to insert: "))
position = int(input("Enter position (0-indexed): "))

my_list.insert(position, number)

print(f"List after insertion: {my_list}")
```

#4  Write a program to rename a key city to a location in the following dictionary. sample_dict = { "name": "Kelly",  "age":25,  "salary": 8000, "city": "New york"}

```python
sample_dict = {
    "name": "Kelly",
    "age": 25,
    "salary": 8000,
    "city": "New york"
}

print("Original dictionary:", sample_dict)

# Get the value associated with the 'city' key
city_value = sample_dict.pop("city")

# Add a new key 'location' with the retrieved value
sample_dict["location"] = city_value

print("Dictionary after renaming 'city' to 'location':", sample_dict)
```
