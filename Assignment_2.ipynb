FizzBuzz

```python
# 1
for number in range(1, 101):
    if number % 3 == 0 and number % 5 == 0:
        print("FizzBuzz")
    elif number % 3 == 0:
        print("Fizz")
    elif number % 5 == 0:
        print("Buzz")
    else:
        print(number)
```

Electricity bill

```python
# 2
units_consumed = int(input("Enter units consumed: "))
total_bill = 0
if units_consumed <= 100:
    total_bill = units_consumed * 5
elif units_consumed <= 200:
    total_bill = (100 * 5) + ((units_consumed - 100) * 7)
else:
    total_bill = (100 * 5) + (100 * 7) + ((units_consumed - 200) * 10)

print(f"Total Bill Amount: ₹{total_bill}")


```

Pallindrome

```python
s = input("enter num or string to be selected")
if s == s[::-1]:
    print("given_num/string is palindrome")
else:
    print("given_num/string is not palindrome")
```
