Design a Mini Banking System that simulates basic banking operations.

The system should support account creation, deposits, withdrawals, and fund transfers between accounts. Each transaction must be recorded and retrievable as a mini statement.

Business rules include minimum balance enforcement, daily withdrawal limits, and rejection of invalid transactions. Data persistence during runtime and menu-based navigation are mandatory.


```python
acc_no = 0
accounts = {}
transactions = {}
daily_withdraw = {}
MIN_BAL = 100
DAILY_WITHDRAW_LIMIT = 50000

while True:
    print("\n--- Mini Banking System ---")
    print("1. Create Account")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Transfer")
    print("5. Mini Statement")
    print("6. Exit")

    try:
        choice = int(input("Enter choice: "))
    except ValueError:
        print("Invalid input. Please enter a number.")
        continue

    if choice == 1:
        name = input("Enter name: ")
        accounts[acc_no] = MIN_BAL
        transactions[acc_no] = [f"Account created with balance {MIN_BAL}"]
        daily_withdraw[acc_no] = 0
        print("Account created. Account No:", acc_no)
        acc_no += 1

    elif choice == 2:
        try:
            a = int(input("Account No: "))
            amt = float(input("Amount: "))
        except ValueError:
            print("Invalid input. Please enter numbers for account and amount.")
            continue

        if a in accounts and amt > 0:
            accounts[a] += amt
            transactions[a].append(f"Deposited {amt}")
            print(f"Amount {amt} deposited to account {a}. New balance: {accounts[a]}")
        else:
            print("Invalid transaction: Account not found or amount is not positive.")

    elif choice == 3:
        try:
            a = int(input("Account No: "))
            amt = float(input("Amount: "))
        except ValueError:
            print("Invalid input. Please enter numbers for account and amount.")
            continue

        if a in accounts and amt > 0:
            if accounts[a] - amt >= MIN_BAL:
                if daily_withdraw[a] + amt <= DAILY_WITHDRAW_LIMIT:
                    accounts[a] -= amt
                    daily_withdraw[a] += amt
                    transactions[a].append(f"Withdrew {amt}")
                    print(f"Amount {amt} withdrawn from account {a}. New balance: {accounts[a]}")
                else:
                    print(f"Withdrawal rejected: Exceeds daily withdrawal limit of {DAILY_WITHDRAW_LIMIT}.")
            else:
                print(f"Withdrawal rejected: Insufficient balance. Minimum balance of {MIN_BAL} required.")
        else:
            print("Invalid transaction: Account not found or amount is not positive.")

    elif choice == 4:
        try:
            f = int(input("From Account: "))
            t = int(input("To Account: "))
            amt = float(input("Amount: "))
        except ValueError:
            print("Invalid input. Please enter numbers for accounts and amount.")
            continue

        if (f in accounts and t in accounts and amt > 0 and
            accounts[f] - amt >= MIN_BAL and
            f != t): # Ensure transfer is not to the same account
            accounts[f] -= amt
            accounts[t] += amt
            transactions[f].append(f"Transferred {amt} to {t}")
            transactions[t].append(f"Received {amt} from {f}")
            print(f"Transferred {amt} from account {f} to {t}.")
            print(f"Account {f} new balance: {accounts[f]}")
            print(f"Account {t} new balance: {accounts[t]}")
        else:
            print("Transfer rejected: Invalid accounts, amount, insufficient balance, or transferring to the same account.")

    elif choice == 5:
        try:
            a = int(input("Account No: "))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue

        if a in accounts:
            print("\nMini Statement for Account:", a)
            for tr in transactions[a]:
                print(tr)
            print("Current Balance:", accounts[a])
            print("Today's Total Withdrawal:", daily_withdraw[a])
        else:
            print("Account not found.")

    elif choice == 6:
        print("Thank you for using the bank.")
        break

    else:
        print("Invalid choice. Please select a valid option (1-6).")
```

```python
courses = {}
students = {}
MAX_CREDITS = 20

while True:
    print("\n--- University Course Registration System ---")
    print("1. Add Course")
    print("2. Register Student")
    print("3. Enroll Student in Course")
    print("4. Student Summary Report")
    print("5. Course Summary Report")
    print("6. Exit")

    choice = input("Enter your choice: ")

    # 1. Add Course
    if choice == "1":
        course_id = input("Enter course ID: ")
        course_name = input("Enter course name: ")
        credits = int(input("Enter course credits: "))
        seats = int(input("Enter seat limit: "))

        courses[course_id] = {
            "name": course_name,
            "credits": credits,
            "seats": seats,
            "enrolled": 0
        }

        print("Course added successfully.")

    # 2. Register Student
    elif choice == "2":
        student_id = input("Enter student ID: ")
        student_name = input("Enter student name: ")

        students[student_id] = {
            "name": student_name,
            "courses": [],
            "total_credits": 0
        }

        print("Student registered successfully.")

    # 3. Enroll Student
    elif choice == "3":
        student_id = input("Enter student ID: ")
        course_id = input("Enter course ID: ")

        if student_id not in students:
            print("Student not found.")
        elif course_id not in courses:
            print("Course not found.")
        else:
            student = students[student_id]
            course = courses[course_id]

            if course_id in student["courses"]:
                print("Student already enrolled in this course.")
            elif course["enrolled"] >= course["seats"]:
                print("No seats available in this course.")
            elif student["total_credits"] + course["credits"] > MAX_CREDITS:
                print("Credit limit exceeded.")
            else:
                student["courses"].append(course_id)
                student["total_credits"] += course["credits"]
                course["enrolled"] += 1

                print("Student enrolled successfully.")

    # 4. Student Summary Report
    elif choice == "4":
        for sid, info in students.items():
            print(f"\nStudent ID: {sid}")
            print(f"Name: {info['name']}")
            print(f"Enrolled Courses: {info['courses']}")
            print(f"Total Credits: {info['total_credits']}")

    # 5. Course Summary Report
    elif choice == "5":
        for cid, info in courses.items():
            print(f"\nCourse ID: {cid}")
            print(f"Course Name: {info['name']}")
            print(f"Credits: {info['credits']}")
            print(f"Seats Filled: {info['enrolled']}/{info['seats']}")

    # 6. Exit
    elif choice == "6":
        print("Exiting system. Goodbye!")
        break

    else:
        print("Invalid choice. Try again.")
```
